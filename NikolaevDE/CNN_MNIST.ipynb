# Метод обратного распространения ошибки

Рассмотрим двухслойную нейронную сеть:


Метод обратного распространения ошибки определяет стратегию изменения параметров сети $w$ в ходе обучения с использованием градиентных методов оптимизации. Градиентные методы на каждом шаге уточняют значения параметров:

$$ \large {w_{k+1} = w_{k} - \lambda p(w) } $$

где $w_k$ - это синаптические веса на $k$-ой итерации, а $\lambda$ - скорость обучения, причем $ \lambda \in (0;1) $

$p(w)$ - соответствует направлению в многомерном пространстве параметров

Классический метод обратного распространения ошибки определяет величину $p(w)$, как антиградиент функции ошибки на $k$-итерации метода:

$$\large { p(w) = -\nabla E(w_{k}) } $$

Функция ошибки для **стохастического градиентного спуска** определяется так:

$$ \large { E(w) = \frac{1}{L}\sum_{k=1}^{L} \left\{ \frac{1}{2}\sum_{j=1}^{M}(y_j^k - u_j^k  )^2   \right\} } $$ 

$u_j^k$ выходной сигнал $j$-ого нейрона на слое $k$:

$$ \large { u_j = \phi^{(2)}(\sum_{s=0}^{K}{ w_{js}^{(2)} v_s }) = \phi^{(2)}(\sum_{s=0}^{K}{ w_{js}^{(2)} \phi^{(1)}(\sum_{i=0}^{N}{w_{si}^{(1)}{x_i}}) }) } $$


<!-- $$ E(w) = \frac{1}{2}\sum_{j=1}^{M}{( y_j - \phi^{(2)} (\sum_{s=0}^{K}w_{js}^{(2)} \phi^{(1)} (\sum_{i=0}^{N}w^{(1)}_{si}x_i) )  )^2} $$ -->

Следовательно, приведем формулы для вычисления градиента:

$$ \large { E(w) = \frac{1}{2}\sum_{j=1}^{M}(y_j^k - u_j^k  )^2 } $$ 

$$ \large{  E(w) = \frac{1}{2}\sum_{j=1}^{M}(y_j^k - \phi^{(2)}(\sum_{s=0}^{K}{ w_{js}^{(2)} \phi^{(1)}(\sum_{i=0}^{N}{w_{si}^{(1)}{x_i}}) })  )^2 } $$ 

 $$ \large { \frac{\partial E}{\partial w_{js}^{(2)}} = -(y_j - u_j) \frac{d \phi^{(2)}(g_j)}{d g_j}v_s } $$

$$ \large { \frac{\partial E}{\partial w_{js}^{(1)}} = -\sum_{j=1}^{M}(y_j - u_j) \frac{d \phi^{(2)}(g_j)}{d g_j}w_{js}^{(2)} \frac{d \phi^{(1)}(f_s)}{d f_s} x_i } $$

$$ \large { f_s = \sum_{i=0}^{N}w_{si}^{(1)}x_i  \space\space\space\space\space\space\space    g_j = \sum_{s=0}^{K}{w_{js}^{(2)}v_s} }$$

В формулах фигурируют функция активации $\phi(x)$ и ее производная $\phi'(x)$. В нашем случае эту функция выбрана ReLU, соответственно определим её и производную:

```python
def relu(x):
    xx = x.copy()
    
    for i in range(len(xx)):
        for j in range(len(xx[i])):
            xx[i][j] = max(0, xx[i][j])
            
    return xx

def derivative_relu(x):
    xx = x.copy()
    
    for i in range(len(xx)):
        for j in range(len(xx[i])):            
            xx[i][j] = float(xx[i][j] > 0)

    return xx
```

Наивная реализация является очень медленной, поэтому переопределим их более оптимизированными версиями:

```python
def relu(x):
    return np.maximum(x, 0)

def derivative_relu(x):
    return np.where(x > 0.0, 1, 0)
```

В качестве функции активации на **последнем слое** рекомендуется выбирать функцию **softmax**:

$$ \large{ \phi(u_j) = \frac{e^{u_j}}{\sum_{i=1}^{M}{e^{u_i}}} } $$

Её программная реализация:

```python
def softmax(u):
    shift = np.max(u, axis=u.ndim - 1, keepdims=True)
    u = np.exp(u - shift)
    denominator = np.sum(u, axis=u.ndim - 1, keepdims=True)

    return u / denominator
```

В качестве функции ошибки для **задачи классификации**, где каждый входной объект относится **только к одному классу**, выбирается **кросс-энтропия**:

$$ \large { E(w) = - \frac{1}{L} \sum_{k=1}^{L}\sum_{m=1}^{M} y_m^k \space ln(u_m^k) } $$

где $y_j^k = 1$ <=> $x^k$ принадлежит классу ${j}$, иначе ${y_j^k = 0}$

Соответственно, определим функцию:

```python
def compute_cross_entropy(y, y_predicted):
    return np.mean(-np.sum(y * np.log(y_predicted), axis=1))
```

Точность классификации **top-K** определяется так:

$$ \large { topK = \frac{1}{S}{\sum_{j=1}^{S} { {1}_{\{ i_1^j, i_2^j, ..., i_K^j  \}} } ^{(l_j)}  } } $$

- $ \normalsize { \{ i_1^j, i_2^j, ..., i_K^j  \} \in \{ 1,2,...,N \} }$
- $ \normalsize { p^j_{i_1^j}, p^j_{i_2^j},..., p^j_{i_k^j}  - K}$ наибольших достоверностей
- $ \normalsize {l_j}$ - класс, которому принадлежит изображение $\large{I_j}$
- $ \normalsize { {1_{\{ i_1^j, i_2^j, ..., i_K^j\}}}^{(l_j)} } $ - индикаторная функция

Определим такую функцию, в нашем случае, **top-1**:

```python
def compute_accuracy(y, y_predicted):
    return np.mean(np.argmax(y, axis=1) == np.argmax(y_predicted, axis=1))
```

## Этапы обучения
### Инициализация синаптических весов сети
Случайным образом из некоторого распределения.

### Прямой  проход:
1. Вычисление значений выходных сигналов нейронов всех слоев $\large { v_s } $ и $\large{ u_j }$
2. Вычисление значений производных функций активации на каждом слое сети $\Large {\frac{d \phi^{(2)}(g_j)}{d g_j} } $ и $ \Large{ \frac{d \phi^{(1)}(f_s)}{d f_s} }$

### Обратный  проход:
1. Вычисление значений целевой функций и её градиента $ \Large{ \frac{\partial E}{\partial w_{js}^{(2)}} } $ и $\Large {\frac{\partial E}{\partial w_{js}^{(1)}} }$

2. Корректировка синаптических весов:
$$\large { w_{k+1} = w_{k} - \lambda \nabla E(w_{k}) } $$

Таким образом, определим методы прямого и обратного прохода как **forward** и **backward** соответственно:

```python
class Network():
    def __init__(self, input_neurons_size, hidden_neurons_size, out_neurons_size, learning_rate):
        self.input_neurons_size  = input_neurons_size
        self.hidden_neurons_size = hidden_neurons_size
        self.out_neurons_size    = out_neurons_size
        self.learning_rate       = learning_rate
        
        mean = 0.0
        stdw = pow(hidden_neurons_size, -0.5)
        stdb = pow(hidden_neurons_size, -0.5)
        
        # первый слой
        self.w0 = np.random.normal(0.0, stdw, size=(input_neurons_size, hidden_neurons_size))
        self.b0 = np.random.normal(0.0, stdb, size=(hidden_neurons_size))
        
        # второй слой
        self.w1 = np.random.normal(0.0, stdw, size=(hidden_neurons_size, out_neurons_size))
        self.b1 = np.random.normal(0.0, stdb, size=(out_neurons_size))

    # Прямой проход
    def forward(self, x):
        x = np.matmul(x, self.w0) + self.b0
        
        self.w0_1 = x.copy()
        
        x = relu(x)
        
        self.w1_1 = x.copy()

        x = np.matmul(x, self.w1) + self.b1
        x = softmax(x)
        
        return x
    
    # Обратный проход
    def backward(self, x, y, y_predicted):
        e1 = y_predicted - y
        e0 = np.matmul(e1, self.w1.T) * derivative_relu(self.w0_1)

        delta = 1.0 / len(x)

        self.w0 = self.w0 - self.learning_rate * delta * np.matmul(x.T, e0)
        self.b0 = self.b0 - self.learning_rate * delta * np.sum(e0, axis=0)
        
        delta = 0.925 / len(x)

        self.w1 = self.w1 - self.learning_rate * delta * np.matmul(self.w1_1.T, e1)
        self.b1 = self.b1 - self.learning_rate * delta * np.sum(e1, axis=0)
        
    def training_iteration(self, x, y):
        y_predicted = self.forward(x)
        
        self.backward(x, y, y_predicted)
        
    def validate(self, x, y):
        y_predicted = self.forward(x)
        
        loss = compute_cross_entropy(y, y_predicted)
        accuracy = compute_accuracy(y, y_predicted)
        
        return loss, accuracy
```

Определим процедуру обучения:

```python
def train(network, train_x, train_y, epochs=20, batch_size=8):
    
    table_data = []
    
    start = datetime.now()

    for epoch_index in range(epochs):
        ########################################################
        
        epoch_start = datetime.now()
        
        for offset in range(0, len(train_x), batch_size):

            batch_x = train_x[offset:offset + batch_size]
            batch_y = train_y[offset:offset + batch_size]

            network.training_iteration(batch_x, batch_y)
        
        epoch_end = datetime.now()
        
        ########################################################
        clear_output(wait=True)
        
        loss, accuracy = network.validate(train_x, train_y)

        table_data.append([
            epoch_index,
            '{:1.8f}'.format(loss), 
            '{:1.8f}'.format(accuracy),
            '{:1.6f} сек'.format((epoch_end - epoch_start).total_seconds())
        ])

        print(tabulate(table_data, headers=["№", "Ошибка", "Точность","Время эпохи"]))

    end = datetime.now()
    time = (end - start).total_seconds()

    print('\nВремя обучения {} сек'.format(time))
```

*Контрольный набор параметров*: размер пачки может меняться от 8 до 64 изображений (в зависимости от доступного объема памяти), скорость обучения составляет 0.1, количество скрытых нейронов – 300, количество эпох – 20.

```python
input_neurons_size = train_x.shape[1]
hidden_neurons_size = 300
learning_rate = 0.1
class_count = 10
batch_size = 32
epochs = 20

network = Network(input_neurons_size, hidden_neurons_size, out_neurons_size = class_count, learning_rate = learning_rate)

train(network, train_x, train_y, epochs, batch_size = batch_size)
```

```python
loss, accuracy = network.validate(test_x, test_y)

print("Проверка на тестовом наборе данных\n")

print(tabulate([[0, '{:1.8f}'.format(loss), '{:1.8f}'.format(accuracy)]], headers=["№", "Ошибка", "Точность"]))
```

# CNN MNIST PyTorch

```python
import torch
import torchvision
import torchvision.datasets
import torchvision.transforms
import torch.nn.functional
import torch.utils.data
import torch.nn as nn
import numpy as np
import warnings
import matplotlib.pyplot as plt
import os
import math
import time
from matplotlib import pyplot as plot
from torch.autograd import Variable
warnings.filterwarnings('ignore')
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
class Torch_NN(nn.Module):
    def __init__(self, input_neurons_size, hidden_neurons_size, out_neurons_size, learning_rate):
        super().__init__()
        
        self.l1   = nn.Linear(input_neurons_size, hidden_neurons_size)
        self.l2   = nn.Linear(hidden_neurons_size, out_neurons_size)
        self.relu = nn.ReLU()
        self.loss = nn.CrossEntropyLoss()
        
        self.optimizer = torch.optim.SGD(self.parameters(), lr = learning_rate) 
        
    def forward(self, x):
        x = self.l1(x)
        x = self.relu(x)
        x = self.l2(x)
        x = torch.nn.functional.log_softmax(x)
        
        return x

    def backward(self, x, y, y_predicted):
        loss = self.loss(y_predicted, y)
        self.optimizer.zero_grad() 
        loss.backward() 
        self.optimizer.step()

    def training_iteration(self, x, y):
        x = Variable(torch.tensor(x))
        y = Variable(torch.tensor(y))
        
        y_predicted = self(x)
        
        self.backward(x, y, y_predicted)
        
    def validate(self, x, y):
        x = Variable(torch.tensor(x))
        y = Variable(torch.tensor(y))
        
        y_predicted = self(x)
        
        loss     = self.loss(y_predicted, y).data
        accuracy = compute_accuracy(y.detach().numpy(), y_predicted.detach().numpy())
        
        return loss, accuracy
```

Проведем точно такую же процедуру обучения с такими же параметрами, как и в предыдущем разделе:

```python
network = Torch_NN(input_neurons_size, hidden_neurons_size, out_neurons_size = class_count, learning_rate = learning_rate)

train(network, train_x, train_y, epochs, batch_size = batch_size)
```
